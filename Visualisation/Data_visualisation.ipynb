```python
import pandas as pd
import numpy as np

import seaborn as sn
import matplotlib.pyplot as plt
```

### 1. Import data

```python
train_data = pd.read_pickle("../Preprocessing/Data/train_processed.pkl")
eval_data = pd.read_pickle("../Preprocessing/Data/eval_processed.pkl")
```

### 2. Correlation matrix

```python
fig, ax = plt.subplots(figsize=(20,20))

corrMatrix = train_data[['retweet_count', 'month', 'week_day', 'day', 'hour', 'user_verified', 'user_followers_count', 'user_statuses_count','user_friends_count', 'user_mentions_count', 'url_count', 'hashtag_count', 'text_length']].corr()
sn.heatmap(corrMatrix, annot=True, ax= ax)

plt.xticks(rotation=45, fontsize=20)
plt.yticks(fontsize=20)

plt.show()
```

### 3. Retweet count distribution

```python
series = train_data['retweet_count']
fig, ax = plt.subplots()
series.hist(ax=ax, bins=160, grid=False, figsize=(12,8), color='#86bf91', zorder=2)

ax.set_title("Distribution of retweets",pad=20,  weight='bold', size=20)
# log-scale to ease visualisation
ax.set_yscale('log')
# Set x-axis label
ax.set_xlabel("Number of retweets", labelpad=10, size=12)
# Set y-axis label
ax.set_ylabel("Number of Tweets", labelpad=10, size=12)

plt.show()
```

### 4. Miscellaneous plots

```python
plot_train_data = train_data.copy()

plot_train_data = plot_train_data.sample(frac = 0.05)

plot_train_data['retweet_count'] += 1

# for log scale
plot_train_data['user_statuses_count'] += 1
plot_train_data['user_followers_count'] += 1
plot_train_data['user_friends_count'] += 1
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, gridspec_kw={'hspace': 0, 'wspace': 0})
ax1.scatter(plot_train_data['url_count'], plot_train_data['retweet_count'], alpha=0.05)
ax2.scatter(plot_train_data['user_mentions_count'], plot_train_data['retweet_count'], alpha=0.05)
ax3.scatter(plot_train_data['hashtag_count'], plot_train_data['retweet_count'], alpha=0.05)
fig.set_figheight(7)
fig.set_figwidth(18)

ax1.set_ylabel("Number of retweets", labelpad=20, weight='bold', size=12)
ax1.set_xlabel("Number of URLs", labelpad=20, weight='bold', size=12)
ax2.set_xlabel("Number of mentions", labelpad=20, weight='bold', size=12)
ax3.set_xlabel("Number of hashtags", labelpad=20, weight='bold', size=12)

plt.ylim(plot_train_data['retweet_count'].min(), plot_train_data['retweet_count'].max())
plt.yscale('log')
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, gridspec_kw={'hspace': 0, 'wspace': 0})
ax1.scatter(plot_train_data['user_statuses_count'], plot_train_data['retweet_count'], alpha=0.05)
ax2.scatter(plot_train_data['user_followers_count'], plot_train_data['retweet_count'], alpha=0.05)
ax3.scatter(plot_train_data['user_friends_count'], plot_train_data['retweet_count'], alpha=0.05)
fig.set_figheight(7)
fig.set_figwidth(18)

ax1.set_ylabel("Number of retweets", labelpad=20, weight='bold', size=12)
ax1.set_xlabel("Number of statuses", labelpad=20, weight='bold', size=12)
ax2.set_xlabel("Number of followers", labelpad=20, weight='bold', size=12)
ax3.set_xlabel("Number of friends", labelpad=20, weight='bold', size=12)

ax1.set_xscale('log')
ax2.set_xscale('log')
ax3.set_xscale('log')

plt.ylim(plot_train_data['retweet_count'].min(), plot_train_data['retweet_count'].max())
plt.yscale('log')
plt.show()
```
